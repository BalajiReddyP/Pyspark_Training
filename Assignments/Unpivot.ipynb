```
# Solved this in Databricks... Presently DataBricks does not support reading excel files (Need to install some libraries).Hence created some sample data manually to test the code.

from pyspark.sql.functions import *

columns = ['Product', 'Plant Name','Sales_18','Sales_19', 'Sales_20', 'Sales_21', 'Sales_22','Inv_18', 'Inv_19', 'Inv_20','Inv_21', 'Inv_22']

data = [('P1','PL1',20000,25000,19000,22000,23000,13000,12000,13000,13500,14000)]

df = spark.createDataFrame(data = data, schema = columns)

df.show()
```

```
unpivot_col = "stack(10,'Sales_18',Sales_18, 'Sales_19',Sales_19, 'Sales_20',Sales_20, 'Sales_21',Sales_21, 'Sales_22',Sales_22,'Inv_18',Inv_18, 'Inv_19',Inv_19, 'Inv_20',Inv_20,'Inv_21',Inv_21, 'Inv_22',Inv_22) as (Metric,Values)"

unPivotDF = df.select('Product','Plant Name', expr(unpivot_col)).where("Values is not null")
unPivotDF.show()
```
